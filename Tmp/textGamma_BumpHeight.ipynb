```python
from scipy.stats import gamma
import numpy as np
from scipy.integrate import quad

mu = 1
sigma = 1
w = 1
k = 0.1 

alpha_gamma = 2 * mu / sigma**2
scale_gamma = k*sigma**2 / (2 * mu)

print(f"alpha_gamma: {alpha_gamma}, scale_gamma: {scale_gamma}")
x = np.linspace(0, 10, 100)

# gamma_pdf = gamma.pdf(x, alpha_gamma, scale=scale_gamma)
gamma_pdf = gamma.pdf(x, alpha_gamma, scale=0.5)
net_pdf = x**(alpha_gamma - 1) * (1 + k * x**2)**(-w / k/ sigma**2)

gamma_pdf = gamma_pdf / np.sum(gamma_pdf)
net_pdf = net_pdf / np.sum(net_pdf)

import matplotlib.pyplot as plt
plt.plot(x, gamma_pdf, label='Gamma PDF')
plt.plot(x, net_pdf, label='Net PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Gamma PDF vs Net PDF')
plt.show()
```
